# Dataset loading examples

```python
from ThreeWToolkit.dataset import ParquetDatasetConfig
```

## Download from figshare if not on local yet

If the dataset is not found at the given `path`, it will be downloaded automatically.

If you use `force_download=True`, the dataset will be downloaded regardless of whether it already exists locally. If it exists locally, it will be replaced by the new one. `force_download` defaults to `False`.


It may take a while to download from figshare (1.79GB) takes about 2-3 minutes to download.

Only dataset v2.0.0 is supported.

```python
# Modify this path to the folder where your dataset is downloaded
dataset_path = "../../dataset"
```

```python
# Load all files, selecting two classes (0 and 4), build() class from the ParquetDatasetConfig config.
ds = ParquetDatasetConfig(path=dataset_path, target_class=[0, 4]).build()
```

Get the number of events:

```python
print(f"Dataset size: {len(ds)}")
```

## Signal and label split

Our dataset object is a list of DatasetOutputs object, found at the Core directory in our toolkit.

Let's see the components of the first element (Each event is a single DatasetOutputs).

This object contains "signal", "label" and a "metadata" field. 

```python
ds[0]
```

What is in the "signal" key?

```python
ds[0].signal
```

Let's see the ids of the classes that we have in our loaded dataset:

```python
set_of_labels = set()
for event in ds:
    set_of_labels = set_of_labels.union(event.label.unique())
set_of_labels
```

## Customizable label selection

Let's reload the dataset ommiting the target column:

```python
ds = ParquetDatasetConfig(path=dataset_path, target_column=None).build()
```

We expect not to have the "label" key.

```python
ds[0].label
```

The signal should contain 'class' column:

```python
ds[0].signal
```

## Event type splitting

Defining 2 types of events: drawn and simulated

```python
event_types = ["drawn", "simulated"]
```

```python
ds = ParquetDatasetConfig(path=dataset_path, event_type=event_types).build()

print("--------------------------------")
print(f"Filtering events of types {event_types}, there are {len(ds)} events")
```

Only real events now:

```python
event_types = ["real"]
ds = ParquetDatasetConfig(path=dataset_path, event_type=event_types).build()

print("--------------------------------")
print(f"Filtering events of types {event_types}, there are {len(ds)} events")
```

------

## Event class splitting

We can select any combination of class events.

First, let's see how many events the whole dataset has.

```python
# If no target class is provided (equivalent to `target_class=None`), all classes are loaded
ds = ParquetDatasetConfig(path=dataset_path).build()

print("--------------------------------")
print(f"Loading all classes, there are {len(ds)} events")
```

Now, let's select only events of one class (4):

```python
# Only loading events from class 4
target_class = [4]
ds_config = ParquetDatasetConfig(path=dataset_path, target_class=target_class).build()

print("--------------------------------")
print(f"Filtering events of class {target_class}, there are {len(ds_config)} events")
```

Of 2 classes (0, 4):

```python
target_class = [0, 4]
ds_config = ParquetDatasetConfig(path=dataset_path, target_class=target_class).build()

print("--------------------------------")
print(f"Filtering events of class {target_class}, there are {len(ds_config)} events")
```

-------

## File list splitting

We can use a list to select loaded events. This is useful for customized train/val/test splitting.

```python
split = [
    "6/SIMULATED_00012.parquet",
    "6/SIMULATED_00049.parquet",
    "0/WELL-00001_20170201010207.parquet",
    "4/WELL-00001_20170316110203.parquet",
]
```

```python
# To get only files in split, we need to set split="list"
ds = ParquetDatasetConfig(path=dataset_path, split="list", file_list=split).build()
print("--------------------------------")
print(f"Loading files in split, there are {len(ds)} events")
```
