# Save/Load model examples:

This notebook shows practical examples for saving and loading a model checkpoint with the model class instance. 

First, we import `ModelRecorder` which is our primary tool for saving and loading models, along with configuration managers for PyTorch (`MLPConfig`) and scikit-learn models (`SklearnModelsConfig`), and a specific scikit-learn model (`ElasticNet`) for our example.

```python
from ThreeWToolkit.utils import ModelRecorder
from ThreeWToolkit.models import MLPConfig, SklearnModelsConfig
from sklearn.linear_model import ElasticNet
```

# Saving and Loading a PyTorch model:

We instantiate a simple Multilayer Perceptron (MLP) built with PyTorch using `MLPConfig`. Then, we create an instance of `ModelRecorder` and save the PyTorch model to `pytorch_model.pt`.

```python
pytorch_model = MLPConfig(output_size=1, hidden_sizes=[10], input_size=10).build()

model_recorder = ModelRecorder()
model_recorder.save_model(pytorch_model, "pytorch_model.pt")
```

To load the saved PyTorch model, we simply call `load_model()` with the path to the `.pt` file. We can print the loaded model to verify its architecture.

```python
pytorch_model = model_recorder.load_model("pytorch_model.pt")
print(pytorch_model)
```

# Saving and Loading a SKLearn model:

Similarly, for scikit-learn models, we configure an `ElasticNet` model using `SklearnModelsConfig`. We then use `ModelRecorder.save_model()` to save the initialized model to `sklearn_model.pkl`.

```python
sklearn_model = SklearnModelsConfig(model_type=ElasticNet).build()

model_recorder = ModelRecorder()
model_recorder.save_model(sklearn_model, "sklearn_model.pkl")
```

To load the scikit-learn model, we again use `load_model()` with the `.pkl` path. The `ModelRecorder` seamlessly reconstructs the `ElasticNet` object which we can verify by viewing its type.

```python
sklearn_model = model_recorder.load_model("sklearn_model.pkl")
sklearn_model
```
